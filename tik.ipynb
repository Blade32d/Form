```python
import tkinter as tk
from tkinter import messagebox
```

```python
root = tk.Tk()
root.title("Registration Form")
root.geometry("300x250")
```

```python
# Labels
tk.Label(root, text="Name").grid(row=0, column=0, padx=10, pady=5)
tk.Label(root, text="Email").grid(row=1, column=0, padx=10, pady=5)
tk.Label(root, text="Age").grid(row=2, column=0, padx=10, pady=5)
tk.Label(root, text="Gender").grid(row=3, column=0, padx=10, pady=5)

# Entries
name_entry = tk.Entry(root)
email_entry = tk.Entry(root)
age_entry = tk.Entry(root)

name_entry.grid(row=0, column=1, padx=10, pady=5)
email_entry.grid(row=1, column=1, padx=10, pady=5)
age_entry.grid(row=2, column=1, padx=10, pady=5)

# Gender Options
gender_var = tk.StringVar()
gender_var.set("Male")  # default value

tk.Radiobutton(root, text="Male", variable=gender_var, value="Male").grid(row=3, column=1, padx=10, pady=5)
tk.Radiobutton(root, text="Female", variable=gender_var, value="Female").grid(row=3, column=2, padx=10, pady=5)
```

```python
def submit_form():
    name = name_entry.get()
    email = email_entry.get()
    age = age_entry.get()
    gender = gender_var.get()

    if name and email and age:
        messagebox.showinfo("Form Submitted", f"Name: {name}\nEmail: {email}\nAge: {age}\nGender: {gender}")
    else:
        messagebox.showwarning("Input Error", "Please fill in all fields.")

tk.Button(root, text="Submit", command=submit_form).grid(row=4, column=1, padx=10, pady=10)
```

```python
root.mainloop()
```
